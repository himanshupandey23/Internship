```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from lxml import html
```

```python
page=requests.get('https://www.nobroker.in/flats-for-sale-in-pune_pune')
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify)
```

```python
flattitle=soup.find_all('a',class_='nb__3CnI6')
flat=[]
for i in flattitle:
    flat.append(i.text)
flat
```

```python
locationflat=soup.find_all('div',class_='nb__2CMjv')
location=[]

for i in locationflat:
    location.append(i.text)
location
```

```python
temp=soup.find_all('div',class_='nb__2NPHR')
temp
```

```python
area_pune=soup.find_all('div',class_='nb__3oNyC')
area=[]
for i in area_pune:
    area.append(i.text)
area
```

```python
emiprice=soup.find_all('div',class_='font-semi-bold heading-6')
temp=[]
for i in emiprice:
    temp.append(i.text)
temp
```

```python
EMI=[]
for i in range(1,len(temp),3):
    EMI.append(temp[i])
EMI
```

```python
price=[]
for i in range(2,len(temp),3):
    price.append(temp[i])
price
```

```python
NObroker=pd.DataFrame({})
NObroker['HouseTitle']=flat
NObroker['Location']=location
NObroker['Area']=area
NObroker['EMI']=EMI
NObroker['Price']=price
NObroker
```
