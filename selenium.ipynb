```python
time.sleep(4)
#now scarpe 10 job 
job_titles=[]
job_location=[]
company_name=[]
full_job_description=[]

jobtitle=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
jobtitle[0:10]

for i in jobtitle:
    job_titles.append(i.text)
job_titles[0:10]
```

```python
location=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]/span[1]')

for i in location:
    job_location.append(i.text)
job_location[0:10]
```

```python
company=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
company[0:10]
for i in company:
    company_name.append(i.text)
company_name[0:10]
```

```python

# scraping the full job-description, for scraping full job description we have to go in each of the jobs separately

urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")]
for url in urls[0:10]:
    try:
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//section[@class='job-desc']/div[1]").text
        description=raw_description.replace("Contact Person","@@@@@")
        description= description.split("@@@@@")
        full_job_description.append(description[0])
    except NoSuchElementException :
        full_job_description.append("---")
```

```python
# creating the dataframe from the scraped data and taking only first 10 jobs
df=pd.DataFrame({"job_title":job_titles[0:10],"company_name":company_name[0:10],"job_location":job_location[0:10],
                "job_description":full_job_description[0:10]})
df
```

```python

import pandas as pd

from selenium import webdriver

from selenium.webdriver.common.keys import Keys

import time
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
url = "https://www.myntra.com/shoes"
driver.get(url)
```

```python
color_filter = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
color_filter.click()
```

```python
price_filter = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_filter.click()
```

```python
url = driver.find_elements_by_xpath('//li[@class="pagination-number"]//a')

# extract all the shoe links from the above search result
shoe_urls=[]
for i in url[:5]:
    shoe_urls.append(i.get_attribute('href'))
```

```python

# code to extract shoe brand details
sbrand = []
for i in shoe_urls:
    driver.get(i)
    try:
        shoe_brand = driver.find_elements_by_xpath('//div[@class="product-productMetaInfo"]//h3')
        for j in shoe_brand:
            sbrand.append(j.text)
    except:
        sbrand.append("---")

# code to extract shoe description
sdescriptn = []
for i in shoe_urls:
    driver.get(i)
    try:
        shoe_descriptn = driver.find_elements_by_xpath('//div[@class="product-productMetaInfo"]//h4')
        for j in shoe_descriptn:
            sdescriptn.append(j.text)
    except:
        sdescriptn.append("---")
description=[]
for i in range(0, len(sdescriptn), 2):
    description.append(sdescriptn[i])


```

```python
sprice = []
for i in shoe_urls:
    driver.get(i)
    try:
        shoe_price = driver.find_elements_by_xpath("//div[@class='product-price']")
        for j in shoe_price:
            sprice.append(j.text)
    except:
        sprice.append("---")
```

```python
len(sprice)
```

```python
print("Lengths of columns:", len(sbrand), len(description), len(sprice))

# creating the data frame now
df = pd.DataFrame({})
df['Shoe Brand']=sbrand[:100]
df['Shoe Description']=description[:100]
df['Shoe Price']=sprice[:100]
df
```

```python
import pandas as pd

from selenium import webdriver

from selenium.webdriver.common.keys import Keys

import time
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
url = "https://www.amazon.in/"
driver.get(url)

search_prod = driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]')
search_prod.send_keys("Laptop")
search_btn = driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]')
search_btn.click()
```

```python
ltitle=[]
lappy_title = driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in lappy_title[:10]:
    ltitle.append(i.text)
ltitle
```

```python
lrating=[]
lappy_rating = driver.find_elements_by_class_name("a-icon-alt")
for i in lappy_rating[:10]:
    lrating.append(i.get_attribute('textContent'))
```

```python
lprice=[]
lappy_price = driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
for i in lappy_price[:10]:
    lprice.append(i.text)
```
