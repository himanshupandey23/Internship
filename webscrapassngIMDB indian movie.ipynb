```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
from lxml import html
```

```python
top100indianIMDB=requests.get("https://www.imdb.com/list/ls009997493/")
soup=BeautifulSoup(top100indianIMDB.content)
```

```python
top100indianIMDB.content
```

```python
print(soup.prettify)
```

```python
indianmovies_name=soup.find_all('h3',class_='lister-item-header')
indianmovies_name
```

```python
top_movies=[]
for i in indianmovies_name:
    top_movies.append(i.text.replace('\n',''))
top_movies
```

```python
IMDB_rating=soup.find_all('div',class_='ipl-rating-star small')
IMDB_rating
```

```python
rating_movies=[]
for i in IMDB_rating:
    rating_movies.append(i.text.strip())
rating_movies
```

```python
year_release=soup.find_all('span',class_='lister-item-year text-muted unbold')
year_release
```

```python
YROFREL=[]
for i in year_release:
    YROFREL.append(i.text.strip())
YROFREL
    
```

```python
topmovies=pd.DataFrame({})
topmovies['TOP100 moviename']=top_movies
topmovies['IMDB rating']=rating_movies
topmovies['Year of release']=YROFREL
topmovies
```

```python
topmovies.to_csv('top100movies.csv',index=False)
topmovies
```
