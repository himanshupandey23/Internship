```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from lxml import html
```

```python
page=requests.get('https://internshala.com/fresher-jobs')
page
```

```python
    soup = BeautifulSoup(page.content, 'html.parser')
    print(soup.prettify)
```

```python
job=soup.find_all('div',class_='heading_4_5 profile')
job
jobtitle=[]
for i in job:
    jobtitle.append(i.text.strip())
jobtitle
len(jobtitle)
```

```python
company=soup.find_all('div',class_='heading_6 company_name')
company
companyname=[]
for i in company:
    companyname.append(i.text.strip().replace('\n',''))
companyname
len(companyname)
```

```python
ctc=soup.find_all('div',class_='item_body')

ctc
ctc_all=[]
for i in ctc:
    ctc_all.append(i.text.replace('\n','').strip())
ctc_all
```

```python
ctc_company=[]
apply_date=[]
for i in range(1,len(ctc_all),3):
     ctc_company.append(ctc_all[i].strip())
ctc_company
len(ctc_company)
```

```python
apply_date=[]
for i in range(2,len(ctc_all),3):
     apply_date.append(ctc_all[i])
apply_date
len(apply_date)
```

```python
intershala=pd.DataFrame({})
intershala['JOBTITLE']=jobtitle
intershala['Company Name']=companyname
intershala['CTC']=ctc_company
intershala['Apply date']=apply_date
intershala
```
